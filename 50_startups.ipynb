```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import influence_plot
```

```python
data=pd.read_csv("50_Startups.csv")
data.head(5)
```

```python
data.info()
```

```python
data=data.rename({'R&D Spend':'RDS','Administration':'ADMS','Marketing Spend':'MKTS'},axis=1)
data
```

```python
data.describe()
```

```python
data.corr()
```

```python
#model building
model=smf.ols("Profit~RDS+ADMS+MKTS",data=data).fit()
model.params
```

```python
model.tvalues,model.pvalues
```

```python
model.rsquared,model.rsquared_adj
```

```python
#ADMS and MKTS found to be insignificant
#Building slr and mlr models
```

```python
slr_adms=smf.ols("Profit~ADMS",data=data).fit()
slr_adms.tvalues , slr_adms.pvalues  
```

```python
slr_mkts=smf.ols("Profit~MKTS",data=data).fit()
slr_mkts.tvalues , slr_mkts.pvalues
```

```python
mlr_adms_mkts=smf.ols("Profit~ADMS+MKTS",data=data).fit()
mlr_adms_mkts.tvalues , mlr_adms_mkts.pvalues
```

```python
# variables have a significant value
```

```python
# Collinearity Check
# Calculate VIF = 1/(1-Rsquared)

rsq_rds=smf.ols("RDS~ADMS+MKTS",data=data).fit().rsquared
vif_rds=1/(1-rsq_rds)

rsq_adms=smf.ols("ADMS~RDS+MKTS",data=data).fit().rsquared
vif_adms=1/(1-rsq_adms)

rsq_mkts=smf.ols("MKTS~RDS+ADMS",data=data).fit().rsquared
vif_mkts=1/(1-rsq_mkts)
```

```python
d1={'Variables':['RDS','ADMS','MKTS'],'Vif':[vif_rds,vif_adms,vif_mkts]}
Vif_df=pd.DataFrame(d1)
Vif_df
```

```python
# no variable have value>20 so no collinearity is present. hence considering all variables in the model.
```

```python
#Residual Analysis
# Test for Normality of Residuals (Q-Q Plot) using residual model (model.resid)

sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-30000))
```

```python
def standard_values(vals) : return (vals-vals.mean())/vals.std()
```

```python
plt.scatter(standard_values(model.fittedvalues),standard_values(model.resid))
plt.title('Residual Plot')
plt.xlabel(' fitted values')
plt.ylabel('residual values')
plt.show() 
```

```python
# Test for errors or Residuals Vs Regressors
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'RDS',fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'ADMS',fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'MKTS',fig=fig)
plt.show()
```

```python
model2.rsquared
```

```python
#model performance increased by 1% aproxx
```

```python
# building a while loop to eliminate outliers till a specified rsquared value is achieved
while model2.rsquared < 0.98:
    for c in [np.max(c)>1]:
        model2=smf.ols("Profit~RDS+ADMS+MKTS",data=data1).fit()
        (c,_)=model2.get_influence().cooks_distance
        c
        np.argmax(c) , np.max(c)
        data1=data1.drop(data1.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
        data1
    else:
        final_model=smf.ols("Profit~RDS+ADMS+MKTS",data=data1).fit()
        final_model.rsquared , final_model.aic
        print("Thus model accuracy is improved to",final_model.rsquared)
```
